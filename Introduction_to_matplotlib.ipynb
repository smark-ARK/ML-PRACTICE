<h3>Plotting Numpy Arrays</h3>

```python
x=np.linspace(0,10,50)
x
```

<h4>
1. Line Plot
</h4>

```python
fig,ax=plt.subplots()
ax.plot(x,x**2)
ax.set(title="Line Plot of x^2",xlabel="values of x",ylabel="y=squares of x")
plt.show()
```

<h4>
2. Scatter Plot
</h4>

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
ax.set(title="Line Plot of x^2",xlabel="values of x",ylabel="y=squares of x")
plt.show()
```

<h4>
3. Bar Plot
</h4>


```python
nut_butter_prices={
    "Almond Buttor":10,
    "Peanut Butter":8,
    "cashew_Butter":12
}

fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Bar Graph Demo",xlabel="Items",ylabel="prices of the Items")
plt.show()
```

<h3>multiple subplots</h3>

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,7))
ax1.plot(x,np.log(x))
ax1.set(title="logarithms",xlabel="values of x",ylabel="log(x)")
ax2.scatter(x,np.cos(x))
ax1.set(title="cos wave",xlabel="values of x",ylabel="cos(x)")
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax1.set(title="prices of nut buutters",xlabel="items",ylabel="prices")
ax4.hist(np.random.randn(1000))
```

```python
heart_disease=pd.read_csv("Datasets/heart-disease.csv")
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.plot(x="age",y='chol',c="target",kind="scatter")
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(x="age",y='chol',c="target",kind="scatter",ax=ax)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax.set(title="Heart disease and cholestrol levels",xlabel="age",ylabel="cholestrol levels")
ax.legend(*scatter.legend_elements(),title="target")
ax.axhline(over_50["chol"].mean(),linestyle="--")
```

```python
fig, (ax0,ax1)=plt.subplots(figsize=(10,10),nrows=2,ncols=1,sharex=True)
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(title="heart diseas and cholestrol level",ylabel="cholestrol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(over_50["chol"].mean(),linestyle="--")

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(title="heart disease and thalach",xlabel="Age",ylabel="Thalach")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(over_50["thalach"].mean(),linestyle="--")
fig.suptitle("Heart Disease Analysis",fontsize=20,fontweight="bold")

plt.show()
```

<h3>Customizing</h3>

```python
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')#setting style theme
fig, (ax0,ax1)=plt.subplots(figsize=(10,10),nrows=2,ncols=1,sharex=True)
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
ax0.set_xlim([50,80])
ax0.set(title="heart diseas and cholestrol level",ylabel="cholestrol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(over_50["chol"].mean(),linestyle="--")

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")
ax1.set_ylim([60,200])
ax1.set(title="heart disease and thalach",xlabel="Age",ylabel="Thalach")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(over_50["thalach"].mean(),linestyle="--")
fig.suptitle("Heart Disease Analysis",fontsize=20,fontweight="bold")

plt.show()
```
